```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.nodes()
```

```python
G.add_edge(1,2)
G.add_edge(4,2)
G.add_edge(3,5)
G.add_edge(2,5)
G.add_edge(9,8)
G.edges()
```

```python
nx.draw(G)
plt.show()
```

```python
nx.draw(G,with_labels=1)
plt.show()
```

```python
Z=nx.complete_graph(10)
Z.nodes()
```

```python
Z.edges()
```

```python
#number of nodes
Z.order()
```

```python
#number of edges
Z.size()
```

```python
nx.draw(Z,with_labels=1)
plt.show()
```
